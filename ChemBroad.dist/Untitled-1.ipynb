```python

core = 9
orbits = [2,8,8,8,8,8,8]
oebits_count = 7

d = draw.Drawing(250, 180, origin='center')
        
#画原子核电荷数
core_text = "+" + str(core)
match len(core_text):
    case 2:
        text_x = -59
    case 3:
        text_x = -63
    case 4:
        text_x = -67

d.append(draw.Circle(-50,0,20,fill='none', stroke='black', stroke_width=2))
d.append(draw.Text(core_text, 17, text_x, 6, fill='black', font_family='Times New Roman'))

#画轨道
for i in range(oebits_count):
    r = 35 + i * 15
    rad = math.asin(7.5 / r) / math.pi * 180
    d.append(draw.Arc(-100, 0, r, 30, rad, cw=False, stroke='black', stroke_width=2, fill='none'))
    d.append(draw.Arc(-100, 0, r, - rad, -30, cw=False, stroke='black', stroke_width=2, fill='none'))
for i in range(oebits_count):
    text = str(orbits[i])
    text_x = -24.5 + i * 15
    match len(text):
        case 1:
            text_x += 0
        case 2:
            text_x += -3
    d.append(draw.Text(text, 15, text_x, 5, fill='black', font_family='Times New Roman'))
    
d.set_pixel_scale(2)
d.save_svg('example.svg')
d
```
